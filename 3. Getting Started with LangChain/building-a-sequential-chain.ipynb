#### Setup Cell

Run the setup cell below to import the libraries required to build the LangChain workflow and initialize the language model.

```python
# Import required libraries
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Initialize the LLM 
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7
)
```

**Note:** Be sure to run the Setup cell above before beginning with the checkpoints!  

#### Checkpoint 1/5: Creating the Explanation Prompt Template

Use the `PromptTemplate` class to create a reusable template named `explanation_prompt` with an instruction and a placeholder that is filled in at runtime. This template will generate a detailed explanation based on a user-provided topic. 

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Prompt to generate a paragraph based on a topic
## YOUR SOLUTION HERE ##
explanation_prompt = PromptTemplate.from_template("Explain {topic} in simple terms.")
```

#### Checkpoint 2/5: Creating the Summary Prompt Template

Using the `PromptTemplate` class, create another template named `summary_prompt` with an instruction and a placeholder that is filled in at runtime. This template will take a detailed explanation from the first chain and summarize it. 

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Prompt to summarize the paragraph based on an explanation
## YOUR SOLUTION HERE ##
summary_prompt = PromptTemplate.from_template("Summarize this explanation: {explanation}")
```

#### Checkpoint 3/5: Building Individual Processing Chains

Create two chains that link the prompt templates, the language model, and a string output parser using the pipe operator `(|)` so each component passes its output to the next: 

- Store the first chain in a variable named `explanation_chain` using the `explanation_prompt` template. 
- Store the second chain in a variable named `summary_chain` using the `summary_prompt` template.

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Create individual chains
## YOUR SOLUTION HERE ##
explanation_chain = explanation_prompt | llm | StrOutputParser()
summary_chain = summary_prompt | llm | StrOutputParser()
```

#### Checkpoint 4/5: Implementing Sequential Chain Execution

A function named `run_sequential_chain` that takes a `topic` parameter is defined to run both chains sequentially and return both results. 

- Use `.invoke()` to run the `explanation_chain` with the `topic` and store the result in a variable named `explanation`. 
- Use `.invoke()` to run the `summary_chain` with the `explanation` result and store it in a variable named `summary`.
  
Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Simple sequential execution function
def run_sequential_chain(topic):
    """Run the chains sequentially and return results"""

    ## YOUR SOLUTION HERE ##
    
    # Generate explanation from topic
    explanation = explanation_chain.invoke({"topic": topic})
    
    # Generate summary from explanation
    summary = summary_chain.invoke({"explanation": explanation})
    
    return {
        "explanation": explanation,
        "summary": summary
    }
```

#### Checkpoint 5/5: Running the Complete Sequential Chain

The `topic` variable is defined to collect the user's input for the explanation.  

Call the `run_sequential_chain()` function by passing the `topic` variable and save the output in a variable named `result`.  

Run the sequential chain to generate an explanation and summary for a chosen topic; the `print()` statements will display both the results. 

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Get user input
topic = input("Enter a topic: ")

# Run the sequential chain
print("\n\033[1m> Running sequential chain...\033[0m")

## YOUR SOLUTION HERE ##
result = run_sequential_chain(topic)

print("\033[1m> Finished!\033[0m")

# Display results
print("\nFull Explanation:\n", result["explanation"])
print("\nSummary:\n", result["summary"])
```
