#### Setup Cell

Run the setup cell below to import the libraries required to set up a simple LangChain workflow. 

- **PromptTemplate** structures prompt using placeholders. 
- **ChatOpenAI** provides access to an OpenAI chat model. 
- **StrOutputParser** converts the model’s response into plain text. 

These imports supply the core components used throughout the workflow. 

**Note:** The OpenAI API key is already configured in this workspace, so no additional setup is needed. When working in an external environment, refer to the [OpenAI API](https://platform.openai.com/docs/quickstart?language=python) documentation for details on configuring your API key. 

```python
# Import required libraries
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

**Note:** Be sure to run the Setup cell above before beginning with the checkpoints! 

#### Checkpoint 1/4: Initializing the LLM

Create an instance of `ChatOpenAI` and store it in a variable named `llm`, setting the `model` argument to `"gpt-4o-mini"` and the `temperature` to `0.7` to balance predictable responses with controlled randomness. 

**Note:** A `temperature` value set close to `0` generates deterministic output, while a value close to `1` creates more varied and creative responses.

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Initialize the LLM
## YOUR SOLUTION HERE ##
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)
```

#### Checkpoint 2/4: Defining a prompt template

Use the `PromptTemplate` class to create a template named `prompt` with an instruction and a placeholder that is filled in at runtime. This template formats the user’s input into a complete prompt that the language model can understand. 

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Define a prompt template
## YOUR SOLUTION HERE ##
prompt = PromptTemplate.from_template("Explain {topic} in simple terms.")
```

#### Checkpoint 3/4: Creating a chain

Create a chain named `chain` that links the prompt template, the language model, and a string output parser using the pipe operator `(|)` so each component passes its output to the next. 

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Create a chain
## YOUR SOLUTION HERE ##
chain = prompt | llm | StrOutputParser()
```

#### Checkpoint 4/4: Executing the chain

The `topic` variable is already defined to collect the user input for the explanation. 

Pass this input to the chain using `.invoke()` so it can generate the response, and store the result in a variable named `response`.

The `print()` statement is used to display the final explanation.

Don't forget to run the cell and save the notebook before selecting "Test Work". Open the "Jupyter Help" toggle at the top of the notebook for more details. 

```python
# Get user input
topic = input("Enter a topic: ")

# Run the chain
## YOUR SOLUTION HERE ##
response = chain.invoke({"topic": topic})

# Print the response
print("\nResponse:\n", response)
```
